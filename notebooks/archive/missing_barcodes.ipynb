```python
import sys
import os
import pandas as pd
import glob 
```

```python
fpath = "../config/fastq_paths.txt"

df = pd.read_csv(fpath)
df['root_dir'] = df['file_path'].apply(lambda x: os.path.dirname(x))
print(f"{df.shape=}")
df
```

```python

for root in df['root_dir'].unique():
    print(root)
    
    file_list = glob.glob(f"{root}/*")
    
    for file in file_list:
        if file not in df['file_path'].values:
            
            if "unclassified" in file:
                continue
                
            if not file.endswith('.fastq'):
                continue
                  
            fid = os.path.basename(file)
            fid = fid.replace(".fastq", "")
            fid = fid.replace("barcode", "b")
            
            print(f"{fid}{file}")

    
```
